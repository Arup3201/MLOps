```python
import pandas as pd
import numpy as np

# Series

# Create a series

arr = np.array([1, 2, 3, 4, 5])

s = pd.Series(arr)

print(s)

print(type(s))

# Set index

s = pd.Series(arr, index=['a', 'b', 'c', 'd', 'e'])

print(s)

# Create a series from a dictionary

d = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}

s = pd.Series(d)

print(s)

# DataFrames

# Create a DataFrame

data = {
    'column1': [1, 2, 3, 4, 5],
    'column2': [10, 20, 30, 40, 50],
    'column3': ['a', 'b', 'c', 'd', 'e']
}

df = pd.DataFrame(data)

print(df)

# Set index

df = pd.DataFrame(data, index=['A', 'B', 'C', 'D', 'E'])

print(df)

# Accessing elements

# Access a column

print(df['column1'])

# Access multiple columns

print(df[['column1', 'column2']])

# Access a row

print(df.loc['A'])

# Access multiple rows

print(df.loc[['A', 'B']])

# Access a specific element

print(df.loc['A', 'column1'])

# Access a specific element using index

print(df.iloc[0, 0])

# Access a specific element using column name

print(df.at['A', 'column1'])

# Access a specific element using column index

print(df.iat[0, 0])

# Data Manipulation

# Add a new column

df['column4'] = [100, 200, 300, 400, 500]

print(df)

# Add a new row

df.loc['F'] = [6, 60, 'f', 600]

print(df)

# Drop a column

df.drop('column4', axis=1,inplace=True)

print(df)

# Drop a row

df.drop('F', axis=0,inplace=True)

print(df)

print(df.describe())




```
