```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple plot')
plt.show()
```

```python
# Customizing the plot

plt.plot(x, y, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=12)
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red')
plt.title('First plot')

plt.subplot(1, 2, 2)
plt.plot(x, y2, color='blue')
plt.title('Second plot')
```

```python
# Bar chart

x = [1, 2, 3, 4, 5]

y = [10, 20, 30, 40, 50]

plt.bar(x, y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar chart')
plt.show()
```

```python
# Pie chart

sizes = [10, 20, 30, 40]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie chart')

plt.show()

```

```python
# Scatter plot

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot')
plt.show()
```

```python
# Histogram

import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color='green', edgecolor='black')
plt.title('Histogram')
plt.show()
```

```python
# Visualizing data from a CSV file

import pandas as pd

data = pd.read_csv('sales.csv')

total_profit_per_month = data.groupby('month_number').total_profit.sum()

total_profit_per_month.plot(kind='line')

plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.title('Total profit per month')

```
