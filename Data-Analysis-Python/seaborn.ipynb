```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sales = pd.read_csv('sales.csv')

sales_month_facewash = sales.groupby('month_number').total_profit.sum()

sns.lineplot(x=sales_month_facewash.index, y=sales_month_facewash.values)
```

```python

sns.barplot(x=sales_month_facewash.index, y=sales_month_facewash.values)
plt.show()
```

```python
sns.histplot(sales_month_facewash, kde=True, bins=10)
```

```python
sns.boxplot(sales_month_facewash)
```

```python
sns.violinplot(sales_month_facewash)
```

```python
sns.kdeplot(sales_month_facewash, fill=True)
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(sales.corr(), annot=True, cmap='coolwarm')
plt.show()
```
